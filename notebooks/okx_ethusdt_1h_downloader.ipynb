# OKX ETH/USDT 1H Downloader

Notebook ini menunjukkan cara mengambil data OHLCV dari bursa OKX dengan `ccxt` dan menyimpannya menjadi file Excel yang dapat digunakan untuk pengujian strategi trading.

````python
import ccxt
import pandas as pd
````

````python
# Inisialisasi objek exchange OKX
exchange = ccxt.okx()
````

````python
# Ambil data OHLCV
data = exchange.fetch_ohlcv(
    symbol='ETH/USDT:USDT',
    timeframe='1h',
    since=exchange.parse8601('2025-01-01T00:00:00Z'),
)

len(data)
````

````python
# Buat DataFrame dengan kolom yang diinginkan
df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df.head()
````

````python
# Simpan ke file Excel
output_path = 'data/OKX_ETHUSDT_1H.xlsx'
df.to_excel(output_path, index=False)
print(f'Data berhasil disimpan ke {output_path}')
````
